

## Introduction

This week I watched a debate called The Great Chatbot Debate: Do LLMs Really Understand? featuring Emily M. Bender and Sbastien Bubeck, moderated by Eliza Strickland. The discussion asked do large language models actually understand anything, or do they just simulate the appearance of understanding? Before class, I was asked to think about three questions:

Do LLMs exhibit understanding in any meaningful sense?

Why or why not?

Why is (or isnt) it important for them to understand?

My personal response to these is that LLMs don't exhibit understanding in any meaningful sense because their knowledge comes from data sets and not lived experiences. However, I don't think it is important for them to understand for simple tasks but is important for topics that have higher stakes.

To test how LLMs would approach this topic, I used Perplexity to generate a detailed summary of the debate. It outlined the main arguments clearly, dividing the perspectives between Benders skepticism and Bubecks optimism.

## What I Learned from Perplexity

Perplexity correctly summarized the arguments.

For example, Perplexity noted that Bender argued that language models do not truly understand anything because they work with patterns of words rather than meaning. They lack grounding in the real world and cannot connect language to actual experiences, intentions, or context. She warned against anthropomorphizing these systems, especially as they begin to be used in sensitive areas like law or therapy. Perplexity also document that Sbastien Bubeck, in contrast, argued that understanding might exist on a spectrum. He suggested that while LLMs do not understand in a human sense, they still demonstrate forms of functional understanding because they can solve complex problems and generate insights that are useful to humans.

Perplexitys tone was neutral and analytical. It captured the ideas well but didnt express any clear stance of its own.

> Me: Whats the main takeaway from the debate?

> Perplexity: The debate does not resolve whether LLMs truly understand, but it makes clear that understanding depends on definition. Bender sees understanding as grounded meaning, while Bubeck views it as a functional capacity that can exist in degrees.

Reading that response, I noticed how carefully Perplexity avoided taking sides. It summarized efficiently but without interpretation. That neutrality made me realize that Perplexity was illustrating Benders point on how it was representing ideas without necessarily understanding them. However, I do recognize that I didn't ask Perplexity to take a side or for it's own opinion. Ironically, this is another reason why I agree with Blender outside of the arguments given.

## Comparing Perplexity and ChatGPT

After prompting and then reading Perplexitys analysis, I asked ChatGPT to reflect on the same question. I wanted to see if another model would express a more interpretive view or give a similar output. I gave ChatGPT the video and a summary of the video's arguments in addition to Perplexity's summary. I then asked:

> Me: Do you think LLMs understand language in any meaningful sense?

> ChatGPT: I think they demonstrate a form of synthetic understanding. They do not experience meaning as humans do, but they represent relationships between words, concepts, and context in ways that resemble understanding. It is a structural rather than experiential process.

This answer struck me as more self-reflective. ChatGPT seemed willing to frame its capabilities as partial understanding, not simply pattern recognition. It described its own limits while still recognizing the complexity of its internal processes. In a way it's correct, as LLMs ability to communicate using words, concepts, and context must demonstrate a level of understanding that is worth acknowledging. After all, my dog, Lulu, would not be able to do this yet I would still stay she understands more than an LLm does.

When I asked both systems why understanding matters, their answers revealed the difference in how they each handled nuance.

> Me: Why does it matter if LLMs understand or not?

> Perplexity: Overestimating AIs understanding can lead to misplaced trust, especially in high-stakes applications. Clarity about what models can and cannot do is essential.

> ChatGPT: Understanding matters because it shapes how humans interact with AI. If users mistake pattern recognition for comprehension, they may rely on the system in contexts where reasoning or empathy is required.

Both were thoughtful, but ChatGPTs framing felt more aligned with Bubecks pragmatic optimism, while Perplexitys answer leaned toward Benders caution.

## My Own View

After reading both analyses and watching the debate myself, I came away thinking that large language models exhibit something that resembles understanding but in a limited sense. They are incredibly effective at identifying patterns and inferring relationships between ideas, they understand language and how to communicate (arguably more effective than humans), but it's almost as if it was too by the "book." What I mean is that they do not connect those ideas to lived experience or intent even thougn they probably mastered how to communicate. Their understanding is linguistic and structural but not experiential. I liken it to Steve Jobs talking about consultants. He said that consulting is like having pictures of various fruits. Consultants can describe it and look at it, but it's still only a picture. The experience of actually tasting it, feeling it, or smelling it isn't there. That's exactly how I feel about whether or not LLMs understand.

At the same time, I think Bubecks perspective has merit. Whether or not LLMs understand the way humans do, they are clearly able to perform useful cognitive tasks that are vital and highly skill-based such as summarizing, reasoning, and even generating new ideas. That level of skill deserves recognition and I'd argure, even better than the average person. However, it does not qualify as true comprehension and you can see that in how it responds/writes. I don't know if an LLM can ever mimic humans but I am sure they can get close. What this experiment taught me is that utility and understanding are not the same, but both matter. LLMs can be powerful tools for reasoning, even without genuine awareness. 

This may be an advantage. I have said this before, AI doesn't need to emulate human intelligence it needs to be artificial intelligence. Otherwise, it's no better than talking to another human being which is great but not anymore special.
