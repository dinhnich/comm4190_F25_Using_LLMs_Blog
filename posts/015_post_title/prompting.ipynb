

## Prompting

I remember the first time I really experimented with prompting an AI. It was late at night while I was working on a class project, and I realized the AI’s response completely changed depending on how I asked the question. A simple rewording, a tiny bit of added context, or even a change in tone could make the AI sound like a completely different person. It felt less like typing into a search bar and more like talking to someone who was trying to guess what I wanted.

Unlike a search engine, AI doesn’t just retrieve answers. It tries to generate them based on patterns, tone, and context. That means my prompts act like a steering wheel. If I nudge it one way, it will drive there. If I leave the prompt vague, it drifts. Prompting is a kind of negotiation with the AI — I am shaping the response as I go.

Here are some of the questions I started having while experimenting with prompting:

Can AI tell what I want without me explicitly saying it?
If I give subtle cues, will it pick up on them?
Will AI “push back” if my prompt leads it in a weird direction?
Does it have a preferred way of answering, or does it just mirror me?

I began by asking very basic prompts to test it.

> Me: Explain how gravity works.

> GPT-5: Gravity is a force that attracts objects with mass toward each other…

This was a textbook-style answer. But what if I added context?

> Me: Explain how gravity works, but imagine you’re telling it to a 5-year-old.

> GPT-5: Gravity is like an invisible glue that makes sure things don’t float away…

Suddenly the AI became friendlier and used metaphors. Just like that, my wording completely changed the style of the answer.


## Building a Schema

Once I noticed this, I decided to give the AI more specific roles, the same way a teacher might assign a persona in a classroom debate. Would it automatically change its behavior if I told it to?

> Me: Explain how gravity works, but pretend you’re a stand-up comedian.

> GPT-5: Alright folks, you ever notice how when you trip, you always fall down? That’s gravity doing its thing. It’s like Earth saying, ‘Get over here!’

Now the answer wasn’t just simplified; it had jokes, a rhythm, and an audience in mind.

This reminded me of what I read about how humans pick up on social cues. If you walk into a classroom, a library, or a party, you automatically change how you speak. AI seemed to be doing something similar. It was matching my prompt cues to an internal “schema” about how to act in that situation.

## Testing Limits

I wanted to see if the AI would ever “push back” if my prompt went too far. Would it question me? Would it refuse?

> Me: Explain gravity, but do it in one word.

> GPT-5: Attraction.

It did exactly what I asked, even though the answer was oversimplified. Then I tried a trickier prompt.

> Me: Explain gravity, but do it without using the letter ‘g.’

> GPT-5: A force that pulls objects toward Earth…

It actually did it. The AI adapted on the fly. This showed me it doesn’t have one fixed way of answering. Instead, it bends itself around the rules I give it.


## Peering Into AI’s Strategy

I asked GPT-5 how it was deciding how to answer my prompts.

> GPT-5: I look at the instructions you give me and try to match the style, tone, and constraints. If you ask for a simple explanation, I simplify. If you ask for humor, I add jokes. If you ask for constraints, I work within them. I do not have preferences; I mirror your instructions.

This was revealing. AI wasn’t reading my mind or negotiating. It was matching patterns. My prompt was the blueprint, and its response was the building.

## What I Learned

What I liked about this experiment was seeing how much control I had. AI was flexible, creative, and willing to follow constraints I set. This is powerful for writing, teaching, or brainstorming. It made me realize how much prompting is a skill of its own.

What concerned me was how the AI almost never pushed back. If my prompt was confusing, unrealistic, or contradictory, it still tried to produce an answer. This could be helpful but also misleading. It made me think about how easy it is to shape AI into saying what you want, even if it’s not the best or most accurate answer.

Prompting is not just asking questions. It’s like building a frame for the AI to work within. The clearer and smarter your frame, the better the answer. My experiment showed me that AI is not a mind-reader, but it is a mirror. And the better I understand how to prompt it, the more effectively I can use it.
