

## Associative Learning

One concept the book explores is the well-known psychological principle of conditioned learning. This refers to the process by which organisms learn to associate one stimulus with another. The most famous example is Pavlov’s experiment, in which a dog was repeatedly exposed to the sound of a buzzer followed by food. After enough pairings, the dog began to salivate at the sound of the buzzer alone.

However, the book also raises a key problem with this type of learning: how does an organism determine which specific cue predicts the unconditioned stimulus (such as food)? In any given situation, multiple cues are present before the stimulus occurs. For instance, when food is brought out, the buzzer might sound, but the temperature, lighting, and background noises are also present. How, then, does the organism learn to associate the buzzer—rather than any of the other stimuli—with the arrival of food?

Yet, this problem isn't solely an issue that humans and living organisms face when going through associative learning. Bennett explains how in machine learning, this is called the credit assignment problem. When something happens, what previous cues do you give credit for predicting it?

Bennet says that our ancient bilaterial ancestors have solved it through four tricks that laid the foundational mechanisms for how neurons make associations. 



## The Four Tricks

Bennett outlines four key strategies that early bilaterian organisms used to form associations:

1. Eligibility Traces
Bennett illustrates this with the example of a slug conditioned to associate a tap on its side with a subsequent shock. The slug will only make this association if the tap occurs about one second before the shock, not after. Such stimuli create short “eligibility trace” windows during which associations can be formed. The general principle is that predictive cues must occur shortly before the event they signal in order to be effective.

2. Overshadowing
Overshadowing occurs when multiple predictive cues are present, but the organism learns to rely on the strongest cue while ignoring weaker ones. For example, if a bright light and a faint odor both precede an event, the bright light will dominate as the predictive signal.

3. Latent Inhibition
Latent inhibition helps organisms distinguish between familiar, irrelevant stimuli and novel, potentially meaningful ones. For instance, the constant sound of cars in the background is unlikely to be treated as a predictive cue—it is filtered out as background noise. This mechanism allows the organism to ask, in effect, “What is different this time?” and focus on unusual or new stimuli when forming associations.

4. Blocking
Blocking occurs when one cue has already been firmly established as predictive of an outcome. Once this association is locked in, new cues that overlap with it are ignored, even if they also precede the same event.



## Connection to AI

In one of my COMM classes, we read Daniel C. Dennet's "Cognitive wheels: The frame problem of AI."

Dennett uses robot parables to illustrate the “frame problem.” The first robot (R1) saves a battery but accidentally drags a bomb with it, failing to anticipate side effects. The second (R1D1) tries to deduce every consequence, but wastes time on factors that were not important (like whether the walls change color), leading to the bomb exploding. The third (R2D1) tries to tag relevant vs. irrelevant facts, but took too long “thinking” that the bomb goes off. These stories show how intelligence requires both knowing which facts to ignore and which to consider, and AI struggles to do this. Dennett argues this is a problem of cognition. Humans appear to effortlessly focus on what matters, yet how we do this is not well understood. 

Dennett situates the frame problem using the struggles early philosophers faced with reasoning and expectation. For instance, Hume suggested that thought worked by associations, but didn't address how a thinker could distinguish which associations were important. AI brings attention to this issue since robots must be fed information about the world and ways to filter all that data. 

There have been proposed strategies like deductive systems that try to track what does or doesn’t change after an action, script and frame theories that allow machines to rely on stereotyped scenarios, and non-monotonic or circumscription logics that aim to model the way humans jump to conclusions, but all have had their faults. Dennett warns against “cognitive wheels,” which are technical fixes that function in narrow cases but do not illustrate what people actually think. He concludes that AI has not solved the frame problem, but it has exposed the mystery of how humans perform rapid, selective common-sense reasoning.

We derive much of our understanding of how the brain works from studying our ancestors, as Bennett’s book illustrates. Yet this still leaves unanswered questions—such as how those ancestors first developed involuntary reflexes. This gap in understanding highlights why artificial intelligence struggles to replicate human intelligence. If we do not fully understand the origins of our own cognitive abilities, it is difficult to engineer them in machines.
