

## Introduction

This week I spent time exploring three notebooks from class that showed different sides of AI. The first focused on image generation, the second on retrieval-augmented generation (RAG) with LangChain, and the third extended that setup with Ollama which us a tool for running large language models locally. ## Notebook 1: Image Generation

The first notebook looked simple at first when we explored it in class. You type in a prompt and the AI creates an image. But it turned out to be doing something smarter behind the scenes. Before generating the image, the notebook had the model rewrite my prompt to make it more specific and detailed.

I started with:

> Me: Make an image of a pig flying on a plane. He is the first officer.

But the AI expanded it into:

> OpenAI: "Visualize a cheerful pig flying a plane, soaring high above the clouds. The pig is donned in a first officer's uniform, complete with the customary hat and epaulettes. The cockpit is bathed in the warm hues of the setting sun, and out of the window, the spectacle of the vast open skies can be seen. The pig exudes competence and joy, for he is the first officer handling the controls of the massive machine, leading it through the tranquil ether."

Then it used that rewritten version to create the image. The rewritten prompt gave the model more concrete visual details to work with. 



I went to GPT-5 and asked why this approach worked so well.

> GPT-5: Most people describe ideas in broad strokes. The AI rewrites your prompt to include sensory and visual details that guide the image model toward a clearer picture.

That explanation made sense. The AI wasn’t taking control away from me but was making my thought more visual. Still, it made me think about how much creativity comes from the machine versus the human. When I asked GPT-5 about that, it said:

> GPT-5: Rewriting is collaboration. You provide the seed, and the model helps it grow into something vivid.

## Notebook 2: LangChain RAG

The second notebook, LangChain RAG, was a shift from creativity to precision. It explored how large language models can pull real information from documents instead of relying only on their built-in training data.

At first, I did not fully understand what “retrieval-augmented generation” meant, so I asked GPT-5 to explain it.

> Me: Can you explain RAG in plain language?

> GPT-5: RAG adds a memory layer. Before answering, the model retrieves relevant chunks of text from an external source. Then it uses that context to generate a more accurate and grounded response.

That explanation helped a lot. It meant the AI was no longer just guessing but was reading before speaking.

I uploaded a few short documents about AI ethics and model transparency and asked:

> Me: What are the main ethical concerns discussed in these documents?

The RAG pipeline searched through the text, retrieved relevant sections, and summarized them like this:

> GPT-5: The documents highlight three key themes: bias in training data, lack of transparency in model decisions, and accountability in high-stakes applications.

I checked the responses and found that it had actually quoted parts of the source files. That was the “aha” moment. RAG was not just talking about AI ethics — it was showing its work. Then I decided to experiment by uploading mixed content. I added one file about AI ethics and another about marketing analytics. When I asked:

> Me: What do these documents say about responsible AI?

It ignored the marketing document and pulled only from the ethics material. That impressed me and showed that retrieval models are not just scanning for keyword but understand context.

Next, I asked something the base model might easily hallucinate on.

> Me: According to these documents, when did the first AI ethics frameworks appear?

The RAG pipeline found a reference to 2019 and cited it directly. When I asked GPT-5 in a fresh session, it gave a vague answer about the “early 2010s.” Seeing those side-by-side made it clear how much grounding matters.


## Notebook 3: LangChain RAG with Ollama

The third notebook took the RAG concept and made it local using Ollama. Setting it up felt like building a small, private version of ChatGPT that lived entirely on my computer.

After connecting Ollama to LangChain and loading some text files, I asked:

> Me: What strategies do these documents suggest for improving AI efficiency?

> Ollama RAG: The documents recommend improving data quality, using smaller specialized models, and reducing redundant computation during training.

It worked perfectly. No cloud connection, no data sharing, just a fast, private model doing retrieval and reasoning locally.

I then asked GPT the signficance of local models.

> Me: Why are people so interested in local models like Ollama?

> GPT-5: Local models give people control. They let users experiment freely, keep their data private, and turn AI from a service into a personal tool.

That answer summed up how I felt. Running something locally made AI feel more mine. Not every model has to live in the cloud.
