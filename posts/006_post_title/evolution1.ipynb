## What is common sense?

There have been numerous definitions of common sense over the years.

Aristotle’s koinè aisthèsis, where humans relied on the 5 senses to perceive “common sensibles” like size and objects, to scholastic thinkers like Avicenna, who claimed common sense to be “internal senses”, to common sense being seen as the judge and censor of the mind. However, Locke helped shift away from common sense as a faculty, and common sense began to mean two things: (1) everyday reasoning capacity (ability to distinguish), and (2) the shared beliefs of a community.

Later, this meaning would give common sense a political purpose in London after 1688. The Toleration Act and the end of the Licensing Act created an environment that fostered differing views. Fear over factionalism led to common sense and politeness being promoted as a means to create cohesion. This idea of a minimal baseline of shared truths anchoring societies and avoiding sectarianism was promoted by writers like Addison and religious sermons.

Yet now, I prefer to use an approach talked about by my professor Duncan Watts and his research into common sense. Rather than treating common sense as an assumed concept, he and his colleagye proposed a framework that defines it in terms of agreement (do people generally say the same thing about a claim) and awareness (do people correctly predict that others will agree).

For example:
> 1. Knowing not to touch a hot stove.
> 2. Looking both ways before crossing the street.
> 3. Understanding that if it’s raining, you might need an umbrella.
Philosophers sometimes point out that what counts as “common sense” can vary across cultures, communities, and time periods—it isn’t always universal.

> Me: Is common sense unique to humans?

> GPT-5: **1. In humans**
Common sense usually means a mix of shared knowledge, intuitive reasoning, and cultural norms.
It’s partly biological (basic survival instincts, like avoiding fire), and partly social (knowing that it’s rude to interrupt).
**2. In non-human animals**
Many animals show practical survival sense — e.g., birds migrating, squirrels storing food, or dogs avoiding danger.
But this tends to be instinct or learned behavior, not the reflective, general-purpose reasoning we call “common sense” in humans.
**3. In machines (like me!)**
AI can simulate a kind of “text-trained” common sense, but it lacks embodied experience, instincts, and cultural grounding.
For example, I can say “Don’t put metal in a microwave” because I’ve seen it in training data, but I don’t feel the danger or discover it by trial and error.
