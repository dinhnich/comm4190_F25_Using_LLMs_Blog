

## Introduction

This week I worked through three notebooks that focused on how large language models reason and act. The first covered the basics of LLMs and understanding, the second explored LangGraph, and the third looked at workflows and agents. Together, they showed how AI can move from generating text to performing structured reasoning and coordinated tasks.

## Notebook 1: LLMs and Understanding

The first notebook looked at what “understanding” actually means for a large language model. We often treat AI like it “knows” things, but this notebook broke that illusion down. It explained that an LLM does not store facts the way humans do. Instead, it predicts text based on patterns it has seen before.

I decided to test this directly.

> Me: What does it mean for you to “understand” something?

> GPT-5: Understanding for me is a process of recognizing relationships between words and concepts. I do not form beliefs, but I can represent meaning through patterns that resemble understanding.

That answer caught my attention. The word resemble felt important. It reminded me that an LLM’s understanding is more like simulation than comprehension. It recognizes structure and meaning but does not experience them.

I tried pushing it a bit further.

> Me: So you are not actually aware of what you are saying?
    
> GPT-5: Correct. Awareness is not part of my process. I can describe what awareness means, but I do not have it.

The clarity of that statement was striking. The notebook showed me that when we talk about AI “understanding,” we really mean pattern-based reasoning. It is a form of linguistic prediction that happens to align with how humans think.

## Notebook 2: Using LangGraph

The second notebook, Using LangGraph, built on that foundation by showing how to organize AI reasoning into visual workflows. LangGraph lets you connect steps in a conversation or process like nodes in a flowchart.

At first, it felt a bit abstract, but once I started linking functions and logic together, it clicked. I was no longer thinking about a single AI response as I was designing a conversation map.

I started with a small example where the model had to summarize text, analyze tone, and then decide on a follow-up action.

> Me: Summarize this article and tell me whether the tone is optimistic or critical.

> LangGraph AI: The article discusses renewable energy growth. The tone is optimistic, emphasizing innovation and progress.

Then the workflow automatically moved to the next step:

> AI Action: Suggest related topics for further reading.

> LangGraph AI: “Sustainable architecture,” “battery storage technology,” and “climate policy transitions.”

It felt like a pipeline of reasoning. The AI performing small, linked tasks that built toward a goal.

When I asked GPT-5 how LangGraph differs from normal prompting, it explained.

> GPT-5: A normal prompt is a single request. LangGraph creates a reasoning structure. It allows models to perform sequences of actions, make decisions, and reuse context efficiently.

That made sense. With LangGraph, I was not just talking to an AI but I was choreographing how it thought.

I tried one more complex flow, where the model needed to search for data, extract key insights, and format them in markdown. It worked smoothly, and I realized that this is how real-world AI systems are being built. They are not one big model answering questions but they are networks of small, task-driven steps.

## Notebook 3: Workflows and Agents

The third notebook, Workflows and Agents, expanded on the idea of structured reasoning by showing how models can act autonomously within a defined system. Agents are AI components that can plan, retrieve information, and perform tasks without direct human supervision, as long as their actions are constrained by the workflow.

I built a small workflow that gathered information and generated a summary report.

> Me: Find recent research on AI ethics and summarize it.

> Agent: Recent research emphasizes transparency, human oversight, and fairness in model development.

Then the agent followed up on its own:

> Agent: Would you like a list of specific papers that support these findings?

The workflow was designed to detect incomplete outputs and generate clarifying steps automatically. When I asked GPT-5 how this process worked, it explained:

> GPT-5: The agent follows predefined logic. If the task is incomplete, it generates a continuation step to maintain progress toward the goal.

This made the system feel organized and reliable. The agent was not being “creative” in the human sense. It was following a structured reasoning process designed to achieve specific objectives.

## Reflection

Working through these notebooks clarified how AI can evolve from text generation to structured reasoning and coordinated action.

The LLMs and Understanding notebook emphasized that models simulate understanding rather than experience it.

The LangGraph notebook demonstrated how structured reasoning allows for multi-step problem solving.

The Workflows and Agents notebook showed how those steps can be automated to form systems that operate independently within limits.

Together, they illustrated a progression from prediction to organization to autonomy. The experiments helped me see AI less as a single model and more as a network of processes that communicate and act together.

At the end of the week, I asked GPT-5 what it thought these three experiments demonstrated.

> Me: What do these notebooks show about where AI is heading?

> GPT-5: They show that AI is moving from conversation to coordination. The next step is not about generating better answers but about managing reasoning and action efficiently.

This summed it up well. Modern AI is not only about what a model can say, but how it can plan, reason, and act in a controlled, transparent way.
