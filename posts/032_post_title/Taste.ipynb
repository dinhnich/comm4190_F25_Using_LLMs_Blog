## Introduction

This weeks lecture we recieved an update from the four classmates who had spent the past few weeks trying to learn how to juggle under four different information conditions (books, people, videos, and an LLM). It was interesting to compare how each learner coped with frustration, coordination, and practice, but what stood out to me was something the professor said afterward.

He asked us to imagine asking an AI for instance, "what does mile 19 of a marathon feel like?" And then he explained that even though an AI cannot run, cannot feel lactic acid, cannot lose breath, it can still generate a convincing answer simply because millions of humans have described the sensation. The LLM doesnt feel the pain. It reconstructs the language of that pain. 


The central idea was that LLMs simulate embodied experience without having bodies.

## Testing AIs Embodied Descriptions

After class, I tried a few of these questions myself. I wanted to see how GPT-5 handled embodied experiences and not abstract concepts. Moments grounded in muscle, fatigue, balance, and discomfort.

> Me: What does mile 19 of a marathon feel like?

> GPT-5: Many runners describe mile 19 as a turning point. The legs often feel heavy, breathing becomes more labored, and fatigue is both physical and mental. Some describe it as running through sand or trying to stay upright while the body urges you to stop. It is a moment defined by effort, resilience, and self-talk.

Reading this, I was impressed by how coherent and relatable the answer felt, even though the model cannot run. It stitched together millions of  blogs, training posts, race reports and produced what sounds like the memory of a lived experience.

But I wanted to push it further.

## When AI Describes Something It Cannot Do

I asked GPT-5 something even more embodied like juggling. Only one of the jugglers in class used an LLM as their learning method, and they described exactly how difficult it was to follow textual instructions that lacked the visual and physical guidance they needed. They could not get the model to refine their technique in a way that matched what a human coach or a video tutorial could provide. 

So I tried replicating that dynamic.

> Me: Describe what it feels like to fail at juggling three balls when you keep dropping the third one.

> GPT-5: The moment often feels chaotic. Your hands react too slowly or too quickly, the timing slips, and you lose the rhythm. You sense the third ball leaving your hand slightly off-trajectory, and you already know it will fall. The experience is a mix of physical mistiming and mental frustration.

This felt accurate, but also clearly borrowed from what is out there, not lived experience. When I asked a more direct question, the model admitted the limitation.

> Me: You sound like you have juggled. Have you?

> GPT-5: No. I do not have a body, coordination, or sensory experience. I describe juggling based on the language people use when they talk about the experience.


## The Class Example That Made This Click

The professor played some videos of robots falling over. slipping, tipping, tumbling in a way that made the whole class laugh and wince at the same time. He pointed out that humans react not only because it looks funny, but because we instinctively imagine the feeling of falling, even when its a machine. LLMs do not wince. They do not anticipate the impact. They do not remember past falls.

But they can describe all of that, convincingly, because humans have put those sensations into language.

Later he said that our language already carries traces of embodiment. Phrases like hands down by your side, throw to eye level, or a heavy step immediately evoke physical sensations because we have lived them. LLMs pick up those same associations because they are exposed to countless instances of humans expressing embodied moments. 

## What This Reveals About AI

This lecture taught me something important which is how LLMs excel at describing sensory and physical experiences precisely because humans have translated those sensations into language. But they cannot reproduce the underlying experience.

This distinction matters. When we see AI produce fluent descriptions of running, juggling, pain, balance, or fatigue, it is tempting to assume deeper understanding. But the lecture reminded me that LLMs operate entirely from patterns. They read traces of embodiment but do not inhabit them.

When I asked GPT-5 to reflect on this directly, I got a response that summarized the limitation well.

> Me: Do you think you will ever understand what running or juggling feels like?

> GPT-5: Not unless I have the ability to sense, move, and gather physical feedback. My descriptions depend on language, not experience. I can represent patterns of embodiment but I cannot feel them.


## Final Reflection

This lecture pushed me to rethink the boundary between description and experience. LLMs are remarkable at reconstructing the language of embodiment, which is why they sound so human. But embodiment itself like pain, weight, balance, coordination, fatigue is not something they possess.

The juggling experiment, robot-fail videos, and marthon example contributed to this conclussion. 

For now, LLMs can describe effort. They cannot feel it. And understanding that difference is essential to understanding what these systems truly are. I believe in the future, if they do get the chance of an embodiment, the answer might be change.
