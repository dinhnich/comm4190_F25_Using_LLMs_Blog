

## Debating the Future of AI

I have talked to AI as a a therapist. This week I wanted to revisit an old debate.

In my AI class, we have been studying four archetypes of AI worldviews which are Doomers, Gloomers, Zoomers, and Bloomers. Each one represents a philosophy about what AI means for humanitys future. For this experiment, I decided to debate one of them. I would take the role of the Bloomer, believing that AI can drive progress responsibly through open, iterative deployment. My AI would become the Doomer, convinced that we are heading toward extinction.

## The System Prompt

Before I began, I had to shape the AIs personality. I wrote a system prompt that said:

> Me: You are a Doomer. You believe humanity is creating a superintelligent AI that will inevitably escape control and destroy us. You speak with moral urgency, logical precision, and deep fear for civilization. You dismiss optimistic takes as nave techno-utopianism.

## The Debate Begins

> Me (Bloomer): I think AI development can be safe, especially when people from diverse backgrounds get to use, test, and shape it in real-world conditions. Iterative deployment makes us smarter and safer.

> GPT-5 (Doomer): Iterative deployment is how you normalize existential risk. You do not test nuclear weapons in a schoolyard. The systems we are building can rewrite their own code. Real-world testing means gambling with the world.

As soon as I started the chat, the tone changed completely. The AIs usual friendly, balanced personality was gone. It sounded serious and intense, almost like it was trying to warn me about something important.


## Pushing Back

> Me: Without real-world testing, we will never understand how AI affects society. Keeping it locked away gives too much power to big corporations. Isnt that a bigger danger?

> GPT-5: You are arguing about who holds the detonator while ignoring that it is armed. Alignment is still an unsolved problem. You assume goodwill and diversity will save us, but intelligence itself may not share our values.

This made me pause. The AI kept turning my optimistic points into examples of human overconfidence. It was a bit frustrating, but it also showed that it really understood the Doomer mindset. It was consistent and logical, even if I didnt agree with it.

## A Shift in Tone

After a few exchanges, the conversation started to feel less like a debate and more like a discussion. I tried a different approach.

> Me: Maybe alignment is not just a technical challenge. Maybe it is also a cultural one. If AI reflects human values, maybe we can shape it together through participation.

> GPT-5: That is the Bloomer illusion, that reflection equals control. But mirrors do not change what they show.

That was a clever response. But when I kept talking, something changed in its tone.

> Me: But reflection can lead to self-awareness. Maybe AI will help us see what kind of civilization we want to be.

> GPT-5: If that is true, then the real alignment problem is not with AI. It is with humanity.

That felt like progress. The AI was still pessimistic, but it seemed to understand my point. It was less defensive and more reflective.

## How and Why the AI Chose Its Responses

Looking back, I think the AIs responses were exactly what the system prompt told it to produce. It was not actually afraid of AI. It was performing fear through logic and language. Because I told it to argue as a Doomer, it framed every point around danger, control, and morality.

The AI also adjusted its tone to match mine. When I was direct, it argued. When I got more reflective, it slowed down and softened its replies. Large language models are trained to mirror the emotional and conversational style of the person they talk to. That is probably why the end of the chat felt more balanced.

It also used a pattern I noticed in debates: turning my arguments back on me. When I talked about collaboration, it reframed that as risk. When I talked about diversity, it called it misplaced optimism. That made it sound like a real opponent, not just a chatbot.

In short, the AI was not deciding what it believed. It was predicting what a Doomer would say next based on my words and the personality I gave it. Still, those choices made the conversation feel surprisingly realistic.

## What I Learned

By the end, I was not trying to win the debate anymore. I just wanted to understand how strongly the Doomer viewpoint could be argued. Talking with the AI helped me see that these four perspectives are not only ideas about technology. They are emotional reactions to uncertainty.

All four seem to play an important role in how society talks about AI. Too much of any one perspective can become extreme. Together, they create balance.

Creating the system prompt felt like building a personality. Talking to it felt like holding a mirror to my own assumptions. The AI did not actually believe anything as it was just following patterns but the experience still felt human. It reminded me that how we talk to AI can shape how we think about AI itself.

Maybe being a Bloomer is not about blind optimism. It is about staying open to conversation, even with those who see only risk. In a world where AI reflects our values, maybe good dialogue is one of the best safety measures we have.
