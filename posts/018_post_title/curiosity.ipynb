

## Curiosity

Something that intrigued me about this reading session was how Bennett spoke about Curiosity. He introduces the idea by talking about Montezuma's Revenge and how TD learning algorithms had difficulty beating it compared to other games like Space Invaders. This is because a game like Montezuma's Revenge requires exploration and there isn't an immediate reward that can be reinforced.

To solve this issue, Google's DeepMind developed a new algorithm that added "curiosity" which tweaked the exploitation-exploration dilemma in trial-and-error learning. In other words, reinforcement learning wasn't just about exploiting behaviors that predict lead to rewards but exploring new behaviors.

## Two Opponent Processes

The difference between exploitation and exploration, is that one is for behaviors that were previously reinforced and the other is for behaviors that are new. They oppose each other as exploitation will drive behavior toward known rewards, and exploration will drive toward what is unknown. 

Bennett mentions how in early TD learning algorithms the trade-off was 5% of the time AI systems would do something random which explains why it didn't perform well in a game like Montezuma since there were an infinite number of directions and places you could go.

## An Alternative Approach

The alternative approach to this is making AI systems explicitly curious by rewarding them for exploring new places and doing new things to make surpirse itself the reinforcement. 

The increase novelty created an increased compulsion to explore it. There was an intrinsic motivation to explore new things that made the AI system act more like a human. It wasn't a random action as these AI systems were deliberate, they specifically wanted to go to new places and to do new things. 

Most importantly, no explicit reward was needed for these AI systems to explore. Finding a way to a new room was the reward in it and of itself. This enabled the AI system to pass level one of Montezuma. 




## Exhibiting Curiosity

"A brain designed to learn through reinforcement should also exhibit curiosity." The book mentions how evidence suggests that early vertebrates were the first to become curious and iut is not unique to humans. Bennett connects this to how surprise triggers the release of dopamine even if there is no "real" reward and that most invertebrates are incapable of curiosity. 

Bennett goes on to explain how the emergence of curiosity helps explain gambling which is irrational behavior since it violates Thorndike's law of effect. People still gamble their money despite the expected reward being negative.

B.F. Skinner was the first to identify this when he found out that rats will actually push levers more times if it was on a variable-ratio schedule rather than a fixed-ratio reinforcement. In other words, it was better to have the lever release food only sometimes then have it all the time.

## Gambling

Bennett explains that vertebrates receive an added boost of reinforcement when they encounter something unexpected. Curiosity emerged as organisms evolved to find surprising and novel experiences rewarding, motivating them to seek out and explore new things. This suggests that even if an activity has a negative reward, its novelty alone can still motivate us to pursue it.

Gambling games are deliberately designed to take advantage of this tendency. In these games, the chance of winning is not 0% as it is often around 48%, which is high enough to make winning seem possible yet uncertain enough that a win feels surprising and delivers a rewarding dopamine boost.

## Social Media

TikTok and Instagram feeds exploit this same mechanism. With each scroll, a new post appears, and after an unpredictable number of scrolls, something interesting shows up. Even if you do not want to keep using Instagram, just as gamblers may not want to gamble or addicts may not want to use again, the behavior is subconsciously reinforced, making it difficult to stop.

## Implication

Gambling and social media feeds exploit our ancient preference for surprise, a trait shaped by evolution but poorly adapted to modern technology. Curiosity and reinforcement learning evolved together because curiosity drives the search for new information, which is essential for learning. Early vertebrates that explored, recognized patterns, and remembered useful places were more likely to survive, making curiosity a valuable evolutionary tool.
