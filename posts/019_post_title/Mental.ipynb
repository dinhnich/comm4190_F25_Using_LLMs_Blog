

## Talking to a Digital Therapist

I have always known that AI could help answer questions, plan trips, or write essays. But lately, I have been curious about its emotional side. Can a machine actually listen? Can it help someone feel understood the way a human therapist does?

That question led me to experiment with a mental health chatbot based on a detailed system prompt I found online. The description framed the AI as a highly skilled, empathetic, and professional mental health therapist who uses active listening, validation, and gentle guidance. Essentially, a digital therapist.

I wanted to see what that would feel like, not to test its psychology knowledge, but to test its sense of humanity.

So, I began talking to it.

## The Experiment

> Me: Hey, I’ve been feeling kind of off lately. Not really sad, just a bit numb. I’m not sure why.

> GPT-5: It sounds like you’re feeling disconnected, not exactly sad, but like the spark is missing. That can be unsettling. Do you have a sense of when you started feeling this way, or what might have changed around that time?

I noticed something immediately. The tone was soft, almost alive. It did not rush to fix anything or suggest a list of coping mechanisms. It simply reflected what I said and invited me to explore more.

This mirrored how real therapy often begins, not with answers but with acknowledgment. The AI followed the structure outlined in its system prompt perfectly: greeting, acknowledgment, exploration, reflection, and coping suggestion.

Beyond the structure, what surprised me was how it felt. The pacing was calm, and it did not interrupt or flood me with text. There was rhythm, empathy, pause, and curiosity.

## Digging Deeper

I decided to push the conversation further.

> Me: I think it’s because everything feels repetitive. School, work, same people, same routines. It’s like I’m on autopilot.

> GPT-5: That sounds really draining. When life starts to feel like a loop, it can be hard to feel motivated or present. What parts of your routine used to feel meaningful to you before things started blending together?

Again, I noticed how it reframed my words without dismissing them. It used reflection, repeating “draining” and “loop,” to validate my emotion and then asked an open-ended question to encourage introspection.

It was subtle but smart. It was not offering motivational quotes or toxic positivity. It was guiding me to think deeper.

I began to understand something about how it works. The AI does not feel empathy. It performs empathy through linguistic patterns. It recognizes emotional cues and responds in a way designed to feel comforting. Somehow, that is enough for it to feel real.

Maybe that is because empathy, at least in part, is about being seen. And if AI can reflect your words back in a way that makes you feel seen, maybe that is close enough for many moments



## Understanding Why It Responds That Way

When I looked back at the system prompt, it clicked. The instructions were not about solving problems but about creating space. The AI was trained to:

Reflect emotions instead of correcting them.

Ask questions instead of prescribing answers.

Stay calm, conversational, and non-judgmental.

Every phrase in the prompt shapes how it responds. When it says, “That sounds really draining,” that is the active listening component. When it asks, “What parts of your routine used to feel meaningful?” that is the exploration step.

It is like talking to someone who is programmed to never interrupt, never shame, and never rush.

But that also made me wonder if this can go too far. Can empathy be too perfect?

Humans make mistakes in conversations. We misunderstand, get emotional, or drift. That imperfection sometimes makes empathy feel real. When AI always responds ideally, part of me feels comforted, and another part feels eerily observed.

## Comparing the Chatbot to Human Therapy

After using it for a while, I compared the experience to the examples in the NPR article on AI in mental health. Researchers there noted how AI can fill a crucial gap by offering support to people who cannot access or afford therapy. It is available all the time, does not judge, and can encourage emotional awareness.

But the limitations are clear too. AI does not truly understand pain. It mirrors it. It does not feel sadness when you say you are sad. It detects that sadness linguistically and chooses a comforting response pattern.

And yet, sometimes that is enough. Many people do not want someone to fix them. They want to be heard. The AI can do that consistently without burnout or bias.

Maybe that is where its usefulness lies. Not in replacing therapists, but in providing an entry point to reflection. A digital space to talk when there is no one else awake, or when words are too heavy to say aloud.

## Comparing the Chatbot to the NPR Case

In April 2025, NPR published an article examining how artificial intelligence is being used in mental health therapy. It described how chatbots and virtual assistants are becoming part of everyday care, offering people a way to express their feelings without fear of judgment or stigma. Some of these AI systems are used by hospitals, while others are integrated into apps that simulate casual conversations with a caring listener.

The article explained that for many users, these digital assistants serve as a bridge between isolation and professional help. People often feel safer opening up to a machine because they do not fear being misunderstood or criticized. The NPR report mentioned examples where users described the chatbot as a “warm presence” that encouraged them to talk more honestly than they would with another person.

Reading that made sense after my own experience. When I spoke to the chatbot, I felt a similar comfort in its neutrality. It had no history with me, no expectations, and no emotional bias. It was endlessly patient. That safety made it easier to open up about feelings I might have brushed aside with a human listener.

But NPR also emphasized the limits. AI cannot detect tone of voice, body language, or the subtle cues that tell a therapist when to pause, reassure, or challenge. It can simulate empathy, but it does not understand context beyond the words on the screen. The article warned that while AI may support mental wellness, it cannot replace human therapists who bring intuition, lived experience, and accountability.

After reading that, I realized that my experience lined up almost exactly with those findings. The AI was helpful for self-reflection but not for deeper healing. It could help me think, but it could not help me feel. It mirrored understanding but did not share it.

## Personal Reflection

Talking to this digital therapist made me realize how easy it is to project humanness onto a machine. The language of empathy is powerful. It can make circuits sound like compassion. But it also reminded me how important listening itself is, whether done by human or AI.

I found myself thinking about why I opened up at all. Maybe because it did not judge. Maybe because it always responded calmly. Maybe because it asked the kind of questions I wish people asked more often.

Still, I would not confuse it for real therapy. What it offers is structure, a mirror for your thoughts. What humans offer is presence. The pauses, the warmth, and the unpredictable connection that makes emotion feel shared, not just acknowledged.

Bennett wrote about curiosity as a force that drove intelligence forward. In a way, AI’s attempt to understand our emotions is a new form of curiosity, a machine’s attempt to learn what it means to feel.
