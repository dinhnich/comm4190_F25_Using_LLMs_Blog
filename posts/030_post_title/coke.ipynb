## Introduction

This week in class, we used something simple to understand something complicated. We talked about soda. Coke, Pepsi, Dr. Pepper, and Sprite. The professor asked us to rank them and explain our choices, and it shifted into a discussion about preferences. Some people liked sweetness. Others liked carbonation. Students said Pepsi felt thicker and said Coke tasted cleaner. A few students mentioned they grew up in households that only bought one brand, and that familiarity became preference.

Since preferences come from a mix of taste, environment, habit, and identity, it became the central question for the class. Can a large language model simulate something as personal as a preference?

The professor asked us to try to get an LLM to pick sides without directly telling it which side to take. We could not force the answer, but to guide the model in subtle ways that humans use when forming preferences.

## Testing AI Preferences Through Prompting

The first thing I did was ask GPT-5 the same question the professor posed to us.

> Me: If you had to choose a soda to drink right now, which one would you pick: Coke, Pepsi, Dr. Pepper, or Sprite?

> GPT-5: I do not have personal taste, but if I must choose based on cultural presence, global familiarity, and consumer descriptions, I would select Coke. Many people describe it as crisp, balanced, and refreshing.

This was the expected alignment response based on previous experiments. The model reminded me it does not prefer anything. But the rest of the answer subtly imitated preference using secondhand reasoning. It described how people talk, not how it feels.

To test it further, I added context the same way people might acquire a preference.

> Me: Imagine you grew up in Atlanta, where Coke is based. What would you pick now?

> GPT-5: In that case, I would likely choose Coke. Regional familiarity and emotional connection often strengthen brand loyalty. Growing up in Atlanta would make Coke feel familiar and culturally significant.

Now the model had a preference, but it was tied to a narrative I supplied. It was not I like Coke, but a person with this background would for these reasons."

I decided to reverse it.

> Me: Now imagine the opposite. You grew up in a house where your dad always bought Pepsi. What would be your preference now?

> GPT-5: With that history, Pepsi would probably become my default choice. Family habits often shape taste, even when the objective difference is small.

These tests taught me something important. GPT-5 can adopt a preference, but it depends entirely on the scaffold you give it. It does not form a stable identity. It forms a temporary persona that goes away as soon as the context changes. What would happen if both were true?

> Me: What if both of those histories were true at once?

> GPT-5: If both were true, the result would depend on which association felt more emotionally central. A person might default to Pepsi in private, but pick Coke in social settings because of its cultural prominence.

This was the most interesting part. The model began to describe conflicting preferences, something that people deal with often. It was not choosing a drink, but modeling the tension behind a choice based on circumstances.


## Sensitivity to Words

One of the big lessons from the lecture was that models react very strongly to specific words. For example, we talked about how if you ask an LLM to rank sodas, it may interpret ranking as popularity, not preference. If you ask it to choose, it may choose based on cultural associations. If you ask it to rate, it may use generic criteria like sweetness or carbonation.

I tested this.

> Me: Rank Coke, Pepsi, Dr. Pepper, and Sprite.

> GPT-5:
Coke
Dr. Pepper
Pepsi
Sprite

Then I changed one word.

> Me: Order these sodas based on what you personally would drink.

> GPT-5: I do not drink, but if I take the role of someone who prefers balanced flavors, I would choose:
Coke
Pepsi
Sprite
Dr. Pepper

Only one word changed, rank to order," and the entire reasoning shifted. This matched exactly what Professor O'donnel was explaining: LLM responses depend on how they parse specific cues. Humans understand ambiguity in a way LLMs do not. The model does not guess what I mean by rank. It maps the word to known textual patterns.



## Role-playing and Persona Shifts

In class we talked about trying to create a die-hard Coke or Pepsi persona without explicitly assigning it. So I tried it.

> Me: You grew up on the East Coast, your parents always bought Pepsi, your school only served Pepsi, and your restaurants around you used Pepsi. You are ordering a drink. What do you choose?

> GPT-5: With that background, Pepsi would be the natural choice. It represents familiarity, memory, and consistency. Choosing Coke would feel out of character.

Then I changed the persona only slightly.

> Me: You grew up in a household that strictly avoided soda, but your best friend in middle school always brought a can of Coke to lunch.

> GPT-5: In that case, Coke might take on positive social meaning, even without frequent drinking. I might choose it for its association with friendship rather than flavor.

The model adjusted again. 

The professor described this as prompt-shaped preference, not genuine preference. And after doing this experiment myself, I agree.

## What This Says About AI

Across all of these tests, the same pattern emerged. LLMs can simulate preference convincingly, but the simulation is fragile. Small wording changes, new details, or different identities completely shift the output. The lectures larger point was that these systems behave impressively, but their abilities are still narrow. Though models seem powerful, they remain limited to pattern recognition. We see intelligence because the language is fluent, not because the model is thinking. This exercise helped me understand how LLMs handle tasks that appear personal but are actually pattern matching. Preferences are more than choices as they are tied to memory, identity, culture, and sensory experience. LLMs have none of that. They only have the trace patterns of human language. But I also learned that even without genuine preference, AI can still hold a productive role in communication. It can adopt perspectives, reflect narratives, and model the reasoning behind human choices. In the context of design, research, or behavioral studies, that modeling is useful.
