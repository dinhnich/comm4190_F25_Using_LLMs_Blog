## Introduction

In class this week, Cameron led us through a project where we used an LLM coder to answer a research question based on a TikTok video featuring a debate. What made this project exciting was that I had no idea this was even possible. I had assumed someone would have to manually rate every single comment as pro or anti to determine the overall sentiment. However, Cameron guided us through using conceptual coding, which made the process both simpler and more enjoyable.

This got me thinking about how an LLM would approach such a task if it started from scratch. Would it take the same or a completely different path? Could someone with no technical background use an LLM to analyze what commenters think? Beyond just watching the content, can we meaningfully assess which side the audience supports by analyzing the comments with AI? And what does that reveal about public opinion, bias, and how we interpret sides in a discussion? I think these questions can deepen our understanding of public perception, though likely on a case-by-case basis.

## Why we want to analyze comments

When it comes to any debate, I assume that factors such as the platform, the publisher, the time of year, the framing of the video, and the type of audience who receives the content all influence the comments I will see. For example, on TikTok I usually expect more Gen Z, funny, and quirky comments. Instagram Reels feels completely different, as I tend to see more unhinged, darker, older, and less sensitive comments. YouTube Shorts often features more wholesome comments from Boomers and Gen X users. The same applies to platforms like Reddit, Quora, Discord, Facebook, X, and others.

What worries me about this wide diversity is not that each platform attracts a certain type of audience, but that it limits how much exposure people have to opposing perspectives. If Im only seeing Democratic viewpoints, that might align with what I like and want, but not with what I need to stay truly informed on a topic. Whats worse is that constantly seeing the same opinion can create the illusion that my view is the most widely shared, even when its not.

Lastly, watching the video gave me my own impression of the message and strength of each argyment, but I wondered whether the comments partway confirmed or contradicted my view. Using AI to scan and summarize comments could help answer: Are viewers mostly aligned with the videos argument, or opposed?
Moreover, it raises a concern that if we use an AI to analyze comments as we do for fact checking, what does that say about how we trust AI to reflect human sentiment?



## How AI would perform the analysis

> Chat-GPT: Heres how I would approach it step-by-step. 1. Collect comments from the TikTok post (if possible via API or scraping, respecting platform policies). 2. Pre-process the comments: remove spam, sort by likes, filter for meaningful responses. 3. Use an LLM or sentiment-analysis model to classify each comments stance: e.g., for the videos argument, against, neutral/unclear. 4. Aggregate results: calculate proportions of each class, identify common themes (via keywords or clustering). 5. Interpret: What topics are driving support? Are criticisms clustered around particular points? What tone are people using (angry, unsure, supportive)? 6. Report: Present findings in a clear summary  e.g., Approximately 65% of comments aligned with the videos viewpoint, 20% opposed, 15% neutral. Then provide qualitative insights.

Unsurprisingly, this step-by-step process is similar to what we exactly did in class. Except it did a pre-process to remove spam but we didn't do that for simplicity reasons.


## What the notebook experiments tell us about this method

From the notebooks we have worked on in class, I learned that AI models can handle large amounts of text fairly well, but their accuracy depends heavily on the quality of the input. Context, tone, slang, and irony in comments can make classification much more difficult. I also realized that using multi-step workflows, where the model first identifies a stance and then explains its reasoning, tends to produce better results than asking it to classify everything at once. It is also important to be transparent about what these models can and cannot do. Just because an AI labels a comment as supportive does not necessarily mean the person actually agrees. This is why having a human involved in the process is still essential, to review unclear cases and interpret the more nuanced meanings that AI often misses. 

## Why this matters

Being able to analyze comment sentiment at scale has value. It can help creators understand their audience, help researchers measure public reaction to media, and help companies monitor brand or message reception. But it also carries risks because relying on AI to interpret tone or meaning can mislead if not done carefully. Comments can be sarcastic, culturally coded, or context-dependent as seen with the Basketball example. If the model mis-labels these, conclusions may be flawed.
