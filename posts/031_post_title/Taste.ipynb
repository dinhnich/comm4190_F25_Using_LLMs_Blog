## Introduction

Last week in class, we ran a blind taste test using Coke and Pepsi, but the real experiment was not about soda. It was about whether a large language model can interpret sensory descriptions the same way a human can.

The professor asked each of us to taste two unlabeled cups and write a short, honest description of each drink. Then, using only our text descriptions, we had to ask an LLM to identify which one was Coke and which was Pepsi. If the LLMs are trained entirely on language, how well can they interpret descriptions of taste which issomething they cannot feel?

The lecture emphasized that LLMs have never tasted anything, but they have processed millions of reviews, opinions, and descriptions of sodas, including the recurring phrases such as syrupy, lighter, more carbonated. Can a language model reconstruct a sensory world from linguistic patterns?

## Testing It Myself

After I wrote my own notes about each cup, I opened GPT-5 and tried the experiment on my own terms.

> Me: I just tasted two drinks.
Drink 1 tasted slightly sweeter and thicker.
Drink 2 tasted lighter, more acidic, with more carbonation.
One of the drinks is Coke and one of the drinks is Pepsi. Which one is which based on these descriptions?

> GPT-5: Based on your descriptions, Drink 1 is likely Pepsi and Drink 2 is likely Coke. Pepsi is often described as sweeter and smoother, while Coke is usually experienced as crisper with a sharper acidic bite.

This matched not only my intuition, but also what my classmates observed. Several students described similar adjectives like Pepsi as sweeter or syrupy, Coke as lighter or more carbonated. GPT-5 appeared to rely on the same associations.

But I wanted to see how stable the models reasoning was, considering the lecture emphasized that prompts need to be tested across multiple regenerations. So I refreshed the output.

Three times in a row, it gave the same interpretation. Only when I changed a single adjective did the model change its guess.

## When AI Gets It Wrong

The lecture included examples of LLMs guessing incorrectly. One student described a soda as slightly nutty, and a model guessed root beer instead of Coke or Pepsi. Another student gave the same description to two different models: one classified the drink correctly, the other reversed the order. 

I wanted to replicate that variability, so I tried giving GPT-5 a more ambiguous description.

> Me: Drink 1 tasted watery and mild.
Drink 2 tasted strong and sweet.
Which one is Coke and which one is Pepsi?

> GPT-5: This could go either way. Pepsi is sweeter, but some people perceive Coke as stronger. My guess is that Drink 2 is Pepsi, but the description is not decisive.

This answer was honest but also showed where the limits appear. Without grounding in actual taste, the model can only estimate based on common textual phrasing. And if the phrasing is ambiguous, the models confidence collapses.



## The Lectures Bigger Point: AI Does Not Sense, It Associates

The discussion eventually moved from soda to cognition. Professor O'Donnel explained that LLMs produce convincing sensory interpretations because they have processed enormous volumes of text where humans describe lived experiences. Those associations become templates the AI uses to map words to likely meanings. In other words, the model is not tasting the drink. It is recognizing the language of tasting. Yet the models ability to imitate the reasoning is often good enough to seem like sensing.

## The Uncanny Line Between Simulation and Experience

One of the most interesting points in the lecture was about how embodied cognition differentiates real human taste from what an AI approximates. Humans form preferences from a lifetime of physical, cultural, and emotional associations. AI forms its preferences from text patterns, probabilities, and dataset biases. The professor framed it saying LLMs simulate, but simulation is not experience.

This reminded me of the soda results. GPT-5 could identify Coke and Pepsi from descriptions because it had seen millions of people describe soda. But if I gave it a description that was unusual, personal, or metaphorical, it sometimes guessed incorrectly or contradicted itself. It could produce the appearance of taste, but not taste itself.



## Final Reflection

We rely heavily on our bodies, memories, and contexts to form preferences. LLMs rely entirely on text. And yet, when the descriptions fit common patterns, the AI can often guess the correct soda better than some humans did. The lecture made it clear that this is not evidence of understanding. It is evidence of sophisticated pattern matching. LLMs operate in a world of words, not sensations. When they get something right, it is  because countless humans put their own tastes into language. Exercises like this help clarify what these systems can do and what they fundamentally cannot. GPT-5 can read my soda description and give me a plausible answer, but only I actually tasted it. Only I know what I experienced. And that distinction matters.
